```python
def draw_scatter(x, y,xlabel,ylabel, title=''):
    axis_font = {'fontname':'Times New Roman', 'size':'16'}
    fig = plt.figure(figsize=(6,5))
    ax1 = fig.add_subplot(111)
    ax1.axis([1,9, 1, 9]) 
    ax1.scatter(x, y,s=12, marker='o', color='#0000AA',alpha=0.4)
    plt.axhline(5, color='#000000',linewidth=2)
    plt.axvline(5, color='#000000',linewidth=2)
    plt.xlabel(xlabel,**axis_font)
    plt.ylabel(ylabel,**axis_font)
    plt.title(title)
    plt.show()
```
